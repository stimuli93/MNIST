```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
X = train.drop(['label'],axis=1).values
y = pd.get_dummies(train['label']).values
X = X / 255.0
```

```python
from CNN_tensorflow import CNNTensorflow
cnn = CNNTensorflow()
```

```python
cnn.train(X,y)
```

```python
test = pd.read_csv('test.csv')
X_test = test.values
X_test = X_test/255.0
```

```python
ypred = cnn.sess.run(cnn.predict(),feed_dict={cnn.X_inp:X_test,cnn.keep_prob:0.5})
```

```python
ypred_labels = np.argmax(ypred,axis=1)
```

```python
ypred_labels.shape
```

```python
columns = ['ImageId','Label']
df = pd.DataFrame(data=np.zeros((test.shape[0],2)), columns=columns)
df['ImageId'] = xrange(test.shape[0])
df['ImageId'] += 1
df['Label'] = ypred_labels
df.head()
```

```python
df.to_csv('CNN.csv',index=False)
```
